```python
import random
from datasets import load_dataset
import pandas as pd
```

## Load and check dataset

```python
# Load image metadata from huggingface into pd DataFrame
image_metadata = load_dataset("MAPS-research/ImageGem-images", split="train")
image_metadata = image_metadata.to_pandas()
image_metadata.head()
```

```python
image_metadata.columns
```

```python
# Load model metadata from huggingface into pd DataFrame
model_metadata = load_dataset("MAPS-research/ImageGem-models", split="train")
model_metadata = model_metadata.to_pandas()

# preprocess model metadata
model_metadata['tags'] = model_metadata['tags'].str.split(', ')
model_metadata['trainedWords'] = model_metadata['trainedWords'].str.split(', ')

model_metadata.head()
```

```python
model_metadata.columns
```

## Model medata example usage

###  List all possible base model architectures that LoRAs are trained on

```python
base_models = model_metadata['baseModel'].unique()
base_models
```

### Find LoRA modelversions trained on SD 1.5 with their primary LoRA files

```python
model_primary = model_metadata[(model_metadata['primary']==1)
                                & (model_metadata['baseModel']=='SD 1.5')].reset_index()
model_primary.sample(5)
```

### Find LoRAs with certain tag

```python
# get all tags
all_tags = model_metadata['tags'].explode().unique()
target_tag = random.sample(all_tags.tolist(), k=1)[0]
target_tag
```

```python
models_with_target_tag = model_metadata[model_metadata['tags'].apply(lambda x: target_tag in x)]
models_with_target_tag.head()
```

### Find all modelversion used for generating an image
In some cases, muliple loras may be used to generate a single image

```python
image_loras = image_metadata.groupby('imageId')['modelVersionId'].apply(list).reset_index()
image_loras[image_loras['modelVersionId'].apply(len) > 1].sample()
```

### Find all images and models uploaded by a certain user
Note that image metadata and model metadata table share the same hash table for username

```python
# randomly sample a userHash
user = model_metadata['userHash'].sample(1).item()
user
```

```python
# find all images uploaded by the user
user_upload_images = image_metadata[image_metadata['userHash']==user]
user_upload_images.head()
```

```python
# find all models uploaded by the user
user_upload_models = model_metadata[model_metadata['userHash']==user]
user_upload_models.head()
```

### Find images NOT uploaded by model uploader, for a certain modelversion

```python
while True: # some model versions don't have any third-party images
    # randomly sample a modelversion
    modelversion_id = model_metadata['modelVersionId'].sample(1).item()
    
    # find the uploader of this model
    user = model_metadata[model_metadata["modelVersionId"]==modelversion_id]["userHash"].item()

    # find images NOT uploaded by model uploader
    third_party_images = image_metadata[
        (image_metadata["modelVersionId"]==modelversion_id) & 
        (image_metadata["userHash"]!=user)
    ]

    if len(third_party_images) > 0:
        break 
print('modelVersionId:', modelversion_id)
print('user:', user)
third_party_images.head()
```
